# K-Nearest Neighbor

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.utils import class_weight
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
import pickle
from src.train_test import *
from src.preprocessor import *
from src.assess import *
```

```python
# Read, split, and transform the data

data = pd.read_csv('smog_data_cleaned.csv')
X_train, X_test, y_train, y_test = split(data)
preprocessor = preprocesser()
```

```python
pipe_opt_knn = make_pipeline(preprocessor, KNeighborsClassifier())
pipe_opt_knn = GridSearchCV(pipe_opt_knn, param_grid= {"kneighborsclassifier__n_neighbors": [3,5,7,9,10,13,15, 20, 30]}, cv=3)
pipe_opt_knn.fit(X_train, y_train)
opt_k = pipe_opt_knn.best_params_
print(opt_k)
```

```python
knn_cv = cv_predict_interval(pipe_opt_knn, X_train, y_train)
mean_cv_knn = get_mean_cv(knn_cv)
result_knn = mean_std_cross_val_scores(pipe_opt_knn, X_train, y_train,return_train_score=True,scoring='accuracy')
misClassificationRate_knn = calc_misclass_rate_pred_interval(pipe_opt_knn.predict_proba(X_test),y_test)
test_score_knn = pipe_opt_knn.score(X_test, y_test)
auc_knn = auc_drawer(pipe_opt_knn, X_test, y_train,y_test)
auc_knn = list(auc_knn.values())
roc_knn = roc_auc_score(y_test, pipe_opt_knn.predict_proba(X_test), multi_class='ovr')

cm_knn = confusion_matrix(y_test, pipe_opt_knn.predict(X_test))
cm_knn = cm_knn.diagonal()/cm_knn.sum(axis=1)
cm_knn = cm_knn.tolist()
```

```python
cv_tb = cv_result_table(knn_cv, mean_cv_knn, result_knn['test_score'],result_knn['train_score'])
test_tb = get_test_result_table(misClassificationRate_knn, cm_knn, test_score_knn, auc_knn, roc_knn).round(4)

print(cv_tb)
print(test_tb)
```

```python
with open('workspace_file/03d_1_knn_cv.pkl', 'wb') as f:
    pickle.dump(cv_tb, f)
    
with open('workspace_file/03d_2_knn_test.pkl', 'wb') as f:
    pickle.dump(test_tb, f)
```
