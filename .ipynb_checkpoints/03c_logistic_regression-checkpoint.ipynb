```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.utils import class_weight
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import SequentialFeatureSelector
import re
import pickle
from src.train_test import *
from src.preprocessor import *
from src.assess import *
from src.lr_interaction import *
```

```python
# Read, split, and transform the data

data = pd.read_csv('smog_data_cleaned.csv')
X_train, X_test, y_train, y_test = split(data)
preprocessor = preprocesser()
```

```python
# Make the pipeline and fit the model

w_1 = {0:1,1:1,2:1,3:1,4:1,5:1,6:1,7:1} 
pipe_mlr= make_pipeline(preprocessor, LogisticRegression(multi_class='multinomial',solver='lbfgs',max_iter=10000))
pipe_mlr_weighted = make_pipeline(preprocessor, LogisticRegression(multi_class='multinomial',solver='lbfgs',class_weight = 'balanced', max_iter=10000))
pipe_mlr.fit(X_train, y_train);
pipe_mlr_weighted.fit(X_train, y_train);
```

```python
lr_cv = cv_predict_interval(pipe_mlr, X_train, y_train)
mean_cv_lr = get_mean_cv(lr_cv)
result_lr = mean_std_cross_val_scores(pipe_mlr, X_train, y_train,return_train_score=True,scoring='accuracy')
misClassificationRate_lr = calc_misclass_rate_pred_interval(pipe_mlr.predict_proba(X_test),y_test)
test_score_lr = pipe_mlr.score(X_test, y_test)
auc_lr = auc_drawer(pipe_mlr, X_test, y_train,y_test)
auc_lr = list(auc_lr.values())
roc_lr = roc_auc_score(y_test, pipe_mlr.predict_proba(X_test), multi_class='ovr')

cm_lr = confusion_matrix(y_test, pipe_mlr.predict(X_test))
cm_lr = cm_lr.diagonal()/cm_lr.sum(axis=1)
cm_lr = cm_lr.tolist()
```

```python
cv_result_tb_lr = cv_result_table(lr_cv, mean_cv_lr, result_lr['test_score'],result_lr['train_score'])
test_result_tb_lr = get_test_result_table(misClassificationRate_lr, cm_lr, test_score_lr, auc_lr, roc_lr).round(4)

print(cv_result_tb_lr)
print(test_result_tb_lr)
```

```python
wlr_cv = cv_predict_interval(pipe_mlr_weighted, X_train, y_train)
mean_cv_wlr = get_mean_cv(wlr_cv)
result_wlr = mean_std_cross_val_scores(pipe_mlr_weighted, X_train, y_train,return_train_score=True,scoring='accuracy')
misClassificationRate_wlr = calc_misclass_rate_pred_interval(pipe_mlr_weighted.predict_proba(X_test),y_test)
test_score_wlr = pipe_mlr_weighted.score(X_test, y_test)
auc_wlr = auc_drawer(pipe_mlr_weighted, X_test, y_train,y_test)
auc_wlr = list(auc_wlr.values())
roc_wlr = roc_auc_score(y_test, pipe_mlr_weighted.predict_proba(X_test), multi_class='ovr')

cm_wlr = confusion_matrix(y_test, pipe_mlr_weighted.predict(X_test))
cm_wlr = cm_wlr.diagonal()/cm_wlr.sum(axis=1)
cm_wlr = cm_wlr.tolist()
```

```python
cv_result_tb_wlr = cv_result_table(wlr_cv, mean_cv_wlr, result_wlr['test_score'],result_wlr['train_score'])
test_result_tb_wlr = get_test_result_table(misClassificationRate_wlr, cm_wlr, test_score_wlr, auc_wlr, roc_wlr).round(4)

print(cv_result_tb_wlr)
print(test_result_tb_wlr)
```
